# Description

The mathematical model consists of two components: the landscape and the optimiser. In the thesis, the landscape is chosen as the multifractional Brownian surface. This is generated on [the FracLab toolbox in MATLAB](https://project.inria.fr/fraclab/); please see the FracLab documentation for generating these surfaces. In this notebook, we plot the fractional and multifractional Brownian surfaces from files saved in `data/fractional_brownian_surfaces`. These are used in the review chapter of the thesis.

## Fractional Brownian surfaces

Three fractional Brownian surfaces are used as examples. These are 1024 x 1024 in size and have HÃ¶lder exponents of 0.3, 0.5 and 0.7. 

```python
import numpy as np
from scipy.io import loadmat
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from mpl_toolkits.axes_grid1 import make_axes_locatable

BIGGER_SIZE = 20
plt.rc('font', size=BIGGER_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=BIGGER_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=BIGGER_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=BIGGER_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=BIGGER_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=BIGGER_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title

fig, axs = plt.subplots(1,3, subplot_kw={"projection": "3d"}, figsize=(15,10))
ax1,ax2,ax3 = axs 
fig.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.05, hspace=None)

# H = 0.3
landscape = loadmat("data/fractional_brownian_surfaces/H=0.3/data.mat")["landscape"]
landscape = (landscape - np.min(landscape))/(np.max(landscape) - np.min(landscape))
X = np.linspace(0, landscape.shape[1], landscape.shape[1])
Y = np.linspace(0, landscape.shape[0], landscape.shape[0])
X, Y = np.meshgrid(X, Y)
ax1.plot_surface(X, Y, landscape, cmap=cm.terrain,
                linewidth=0, antialiased=False)
ax1.set_xlabel(r'$x$', labelpad=8)
ax1.set_ylabel(r'$y$', labelpad=12)
ax1.set_xticks([0,500,1000])
ax1.set_yticks([0,500,1000])
ax1.set_zticks([0,0.5,1])
ax1.set_zticklabels([])

# H = 0.5
landscape = loadmat("data/fractional_brownian_surfaces/H=0.5/data.mat")["landscape"]
landscape = (landscape - np.min(landscape))/(np.max(landscape) - np.min(landscape))
X = np.linspace(0, landscape.shape[1], landscape.shape[1])
Y = np.linspace(0, landscape.shape[0], landscape.shape[0])
X, Y = np.meshgrid(X, Y)
ax2.plot_surface(X, Y, landscape, cmap=cm.terrain,
                linewidth=0, antialiased=False)
ax2.set_xlabel(r'$x$', labelpad=8)
ax2.set_ylabel(r'$y$', labelpad=12)
ax2.set_xticks([0,500,1000])
ax2.set_yticks([0,500,1000])
ax2.set_zticks([0,0.5,1])
ax2.set_zticklabels([])

# H = 0.7
landscape = loadmat("data/fractional_brownian_surfaces/H=0.7/data.mat")["landscape"]
landscape = (landscape - np.min(landscape))/(np.max(landscape) - np.min(landscape))
landscape = np.fliplr(landscape)
X = np.linspace(0, landscape.shape[1], landscape.shape[1])
Y = np.linspace(0, landscape.shape[0], landscape.shape[0])
X, Y = np.meshgrid(X, Y)
im3 = ax3.plot_surface(X, Y, landscape, cmap=cm.terrain,
                linewidth=0, antialiased=False)
ax3.set_xlabel(r'$x$', labelpad=8)
ax3.set_ylabel(r'$y$', labelpad=12)
ax3.set_xticks([0,500,1000])
ax3.set_yticks([0,500,1000])
ax3.set_zticks([0,0.5,1])
ax3.set_zticklabels([])

# Colorbar
cmap = plt.get_cmap('terrain')
norm = matplotlib.colors.Normalize(vmin=0,vmax=1)
sm = cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
cbar_ax = fig.add_axes([0.92, 0.34, 0.01, 0.3])
fig.colorbar(sm, ticks=[0,0.25,0.5,0.75,1], label=r'$B_{H}(\mathbf{x})$', cax=cbar_ax, orientation='vertical')

# Labels
fig.text(0.13, 0.625, '(a)')
fig.text(0.39, 0.625, '(b)')
fig.text(0.65, 0.625, '(c)')
fig.text(0.24, 0.26, r'$H=0.3$')
fig.text(0.502, 0.26, r'$H=0.5$')
fig.text(0.767, 0.26, r'$H=0.7$')

fig.savefig("figures/FBSs.pdf", dpi=600, bbox_inches='tight')
plt.show()
```

## Multifractional Brownian surface
Plot of a 1024 x 1024 multifractional Brownian surface and its pointwise HÃ¶lder exponent.

```python
import numpy as np
from scipy.io import loadmat
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.cm as cm

BIGGER_SIZE = 16
plt.rc('font', size=BIGGER_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=BIGGER_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=BIGGER_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=BIGGER_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=BIGGER_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=BIGGER_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title

fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(121, projection='3d')
ax2 = fig.add_subplot(122)
fig.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.7, hspace=None)

# Multifractional Brownian surface
landscape = loadmat("data/multifractional_brownian_surface/landscape.mat")["landscape"]
landscape = (landscape - np.min(landscape))/(np.max(landscape) - np.min(landscape))
X = np.linspace(0, landscape.shape[1], landscape.shape[1])
Y = np.linspace(0, landscape.shape[0], landscape.shape[0])
X, Y = np.meshgrid(X, Y)
ax1.view_init(elev=45., azim=45)
im1 = ax1.plot_surface(X, Y, landscape, cmap=cm.terrain,
                linewidth=0, antialiased=False)
ax1.set_xlabel(r'$x$', labelpad=7)
ax1.set_ylabel(r'$y$', labelpad=7)
ax1.set_xticks([0,500,1000])
ax1.set_yticks([0,500,1000])
ax1.set_zticks([0,0.5,1])
ax1.set_zticklabels([])

cmap = plt.get_cmap('terrain')
norm = matplotlib.colors.Normalize(vmin=0,vmax=1)
sm = cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
cbar_ax = fig.add_axes([0.43, 0.28, 0.01, 0.431])
fig.colorbar(sm, ticks=[0,0.25,0.5,0.75,1], label=r'$B_{H(\cdot)}(\mathbf{x})$', cax=cbar_ax, orientation='vertical')

# Pointwise HÃ¶lder exponent
L = 1024
x = np.linspace(0,1024, 1000)
y = np.linspace(0,1024, 1000)
xx, yy = np.meshgrid(x,y)
H = 0.5 + 0.2 * np.sin(2*np.pi*xx/L) * np.cos(3*np.pi*yy/(2*L))

im2 = ax2.imshow(H, origin='lower', cmap='jet', vmin=0.3, vmax=0.7, extent=[0,1024,0,1024])
ax2.set_xlabel(r'$x$')
ax2.set_ylabel(r'$y$', labelpad=-5)
ax2.set_xticks([0,500,1000])
ax2.set_yticks([0,500,1000])

cmap = plt.get_cmap('jet')
norm = matplotlib.colors.Normalize(vmin=0.300001,vmax=0.7)
sm = cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
cbar_ax = fig.add_axes([0.91, 0.28, 0.01, 0.431])
cb = fig.colorbar(sm, label=r'$H(\mathbf{x})$', cax=cbar_ax, orientation='vertical')
cb.set_ticks([0.300001, 0.4,0.5,0.6,0.7], [0.3, 0.4,0.5,0.6,0.7])

# Labels
fig.text(0.12, 0.7, '(a)')
fig.text(0.52, 0.7, '(b)')

fig.savefig("figures/MBS.pdf", dpi=600, bbox_inches='tight')
plt.show()
```
